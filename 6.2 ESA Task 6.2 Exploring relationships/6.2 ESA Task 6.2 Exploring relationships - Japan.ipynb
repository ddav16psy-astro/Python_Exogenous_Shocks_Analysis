#### Preliminary Interpretation

##### This will merely highlight divergences with the US correlation heatmap results or interesting relationships.

##### Overnight interest rate had moderate-to-strong positive correlations (+0.34-0.55) for all FX pairs with JPY as the counter currency, except USDJPY, while the 10-year interest rate had moderate positive correlations for all such pairs.  The 3-month interest rate only had moderate positive correlations with EURJPY and GBPJPY.

##### Interest rate variables had mostly low correlations with the GDP measures, diverging from the US experience.

##### Interest rate variables (overnight, 3-month but not 10-year) had a moderate negative correlation with the unemployment rate.

##### The primary stock market index log returns had moderate positive correlation with Real GDP but strong negative correlation with GDP per capita in USD, an odd combination.  It had a moderate negative correlation with the VIX, confirming some connection to US equities risk on/risk off appetite.  It had a moderate positive correlation with the 10-year interest rate.

##### The primary stock market index log returns had strong to very strong positive correlations with all FX pairs having JPY as the counter currency, ranging +0.67 to + 0.82 (GBPJPY).  This suggests the local stock market likes a weak Japanese Yen, which makes sense given the country's heavy reliance on exports to drive economic growth.

##### Real GDP revealed a strong negative correlation with the unemployment rate.  Meanwhile, GDP per capita in USD showed little correlation with the unemployment rate, perhaps because the former metric is USD rather than JPY-denominated.

##### GDP per capita in USD revealed moderate-to-very-strong negative correlations (-0.37 for AUDJPY, to -0.87, for USDJPY) with all FX pairs having JPY as the counter currency.  Meanwhile, Real GDP demonstrated the opposite, moderate positive correlations (+0.30 to +0.47 for AUDJPY) for the same pairs except CHFJPY.

##### The VIX variable, even though a US stock market "fear gauge", had a moderate negative correlation (-0.49) with AUDJPY, which is consistent with the view in FX circles that AUDJPY is a representative global "risk on/risk off" sentiment barometer.  In times of crisis, JPY (after CHF) is a safe haven currency, thus AUDJPY normally declines in such environment.

#### Define a custom function to create a grid of histograms for selected variables, with a categorical variable overlay.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math

def shockstack_grid(df, vars_to_plot, hue='shock_name', bins=20, cols=3, figsize_per_plot=(6, 5),
                     save=False, filename=None, filetype='png', output_dir='.'):
    """
    Plots a grid of color-stacked histograms for selected variables using the 'hue' column (default: 'shock_name').

    Parameters:
        df (pd.DataFrame): DataFrame containing the data.
        vars_to_plot (list of str): List of up to 9 numeric variable names to include.
        hue (str): Column name to use for stacked colors (e.g., 'shock_name').
        bins (int): Number of bins for histogram.
        cols (int): Number of columns in the subplot grid (2 or 3 recommended).
        figsize_per_plot (tuple): Size per subplot (width, height).
        save (bool): If True, saves to disk.
        filename (str): If "save" chosen, specify the filename for the output.
        filetype (str) : If "save" chosen, optional other formats include 'pdf', otherwise defaults to 'png'.
        output_dir (str): Path where figures are saved.
    """
    num_vars = len(vars_to_plot)
    rows = math.ceil(num_vars / cols)
    fig, axes = plt.subplots(rows, cols, figsize=(figsize_per_plot[0] * cols, figsize_per_plot[1] * rows), constrained_layout=True)
    axes = axes.flatten() if num_vars > 1 else [axes]
    
    # Define color palette
    categories = df[hue].dropna().unique()
    palette = sns.color_palette("tab10", len(categories))
    color_dict = dict(zip(categories, palette))
    
    for i, var in enumerate(vars_to_plot):
        ax = axes[i]
        # Drop NA rows for variable and hue
        sub_df = df[[var, hue]].dropna()

        # Create binned intervals
        bin_counts = pd.cut(sub_df[var], bins=bins)
        
        # Cross-tabulate
        ctab = pd.crosstab(bin_counts, sub_df[hue])

        # Ensure consistent order
        ctab = ctab[categories]  # match hue order to color_dict

        # Plot stacked bar
        ctab.plot(kind='bar', stacked=True, ax=ax, color=[color_dict[c] for c in categories], edgecolor='black', linewidth=0.3, legend=False)

        ax.set_title(var, fontsize=10)
        ax.set_xlabel("")
        ax.set_ylabel("Count")
        ax.tick_params(axis='x', labelrotation=45)

    # Turn off unused subplots
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')

    # Central legend
    handles = [plt.Rectangle((0, 0), 1, 1, color=color_dict[c]) for c in categories]
    fig.legend(handles, categories, title=hue, loc='upper right', bbox_to_anchor=(0.97, 1.05), fontsize=8)
    
    plt.suptitle("Stacked Histograms by Shock Event", fontsize=14)

    if save:
        if filename is None:
            timestamp = pd.Timestamp.now().strftime("%Y%m%d_%H%M%S")
            filename = f"shockstack_grid_{timestamp}.{filetype}"
        else:
            filename = f"{filename}.{filetype}" if not filename.endswith(f".{filetype}") else filename

        full_path = os.path.join(output_dir, filename)
        plt.savefig(full_path, dpi=300, bbox_inches='tight')
        print(f"âœ… Saved to: {full_path}")
    
    plt.show()
```
